```python
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
from skimage import util, exposure, filters, color, morphology
# homedir = 'Citra Tugas Lab 1/'
```

```python
mask = io.imread('mask.jpg')
plt.figure(figsize=(15, 10))
plt.subplot(1,2,1); plt.imshow(mask)
plt.title('Mask', fontsize=20); plt.axis("off")
plt.show()
```

### a. Tampilkan histogram Citra tsb dengan range intensitas 0 - 255

```python
plt.figure(figsize=(15,10))
mask_flat = util.img_as_ubyte(mask).flatten()
plt.subplot(2,1,1)
plt.hist(mask_flat, 256, range=(0,256))
plt.title('Mask Original', fontsize=20)
plt.show()


```

```python
mask_eq = exposure.equalize_hist(mask)
mask_eq_flat = util.img_as_ubyte(mask_eq).flatten()
plt.figure(figsize=(15, 10))
plt.subplot(2,1,1)
plt.hist(mask_eq_flat, 256, range=(0,256))
plt.title('Histogram Equalization', fontsize=20)
plt.subplot(2,1,2)
plt.imshow(mask_eq)
# plt.axis("off")
plt.show()
```

```python
mn = min(mask.flatten())
mx = max(mask.flatten())
b = int(np.floor(255/(mx-mn)))
mask_cs = (mask - mn) * b
mask_cs_flat = util.img_as_ubyte(mask_cs).flatten()
plt.figure(figsize=(15, 10))
plt.subplot(2,1,1)
plt.hist(mask_cs_flat, 256, range=(0,256))
plt.title('Histogram Contrast Stretching', fontsize=20)
plt.subplot(2,1,2)
plt.imshow(mask_cs)
# plt.axis("off")
plt.show()
```

### d. [15] Coba ceritakan perbedaan apa yang dapat anda temukan antara metode histogram equalization dan contrast stretching

Pada histogram contrast stretching, nilai lebih tersebar merata pada sumbu-x ketimbang histogram equalization. Hal tsb menyebabkan gambar pada contrast stretching memiliki tingkat keabuan yang lebih merata ketimbang gambar yang dilakukan equalization. Nilai pada histogram equalization jauh lebih padat disebelah kiri, sehingga gambar yang telah dilakukan equalization lebih gelap ketimbang gambar yang dilakukan contrast stretching

### a. [10] Gunakan Sobel spatial filter dan Prewitt spatial filter untuk menemukan dan menampilkan horizontal edge dari citra tersebut

```python
mie = io.imread('noodles.jpg')
mie_gray = color.rgb2gray(mie)
edge_sobel_horizontal = filters.sobel_h(mie_gray)
edge_sobel_flat = util.img_as_ubyte(edge_sobel_horizontal).flatten()


plt.figure(figsize=(15, 10))

plt.subplot(1,2,1); plt.imshow(mie)
plt.title('Original', fontsize=20); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(edge_sobel_horizontal)
plt.title('Edge_Sobel_Horizontal', fontsize=20);
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(1,2,1)
plt.hist(edge_sobel_flat, 256, range=(0,256))
plt.title('Histogram Sobel Horizontal', fontsize=20)
plt.show()
```

```python
edge_prewitt_horizontal = filters.prewitt_h(mie_gray)
edge_prewitt_flat = util.img_as_ubyte(edge_prewitt_horizontal).flatten()

plt.figure(figsize=(15, 10))
plt.subplot(1,2,1); plt.imshow(mie)
plt.title('Original', fontsize=20); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(edge_prewitt_horizontal)
plt.title('Edge Prewitt Horizontal', fontsize=20);
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(1,2,1)
plt.hist(edge_prewitt_flat, 256, range=(0,256))
plt.title('Histogram Prewitt Horizontal', fontsize=20)
plt.show()
```

```python
cat = io.imread('cat.jpg')
```

### a. [7.5] Terapkan mean filter pada citra tersebut dengan ukuran kernel 3x3


```python
cat_gray = color.rgb2gray(cat)
cat_fi_mean = filters.rank.mean(cat_gray, selem=morphology.square(3))

plt.figure(figsize=(15, 10))
plt.subplot(1,2,1); plt.imshow(cat,cmap='gray',vmin=0,vmax=255)
plt.title('Noise', fontsize=20); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(cat_fi_mean,cmap='gray',vmin=0,vmax=255)
plt.title('Mean Filter', fontsize=20);
plt.axis("off")
plt.show()
```

### b. [7.5] Terapkan median filter pada citra tersebut dengan ukuran kernel 3x3

```python
cat_fi_median = filters.rank.median(cat_gray, selem=morphology.square(3))

plt.figure(figsize=(15, 10))
plt.subplot(1,2,1); plt.imshow(cat,cmap='gray',vmin=0,vmax=255)
plt.title('Noise', fontsize=20); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(cat_fi_median,cmap='gray',vmin=0,vmax=255)
plt.title('Median Filter', fontsize=20);
plt.axis("off")
plt.show()
```
