```python
from skimage import io, color
from skimage.morphology import disk, square, diamond, dilation, erosion, closing, opening, black_tophat, white_tophat
import matplotlib.pyplot as plt
```

## 1. [25] Shape Separation

### a. [15] Lakukan berbagai percobaan manipulasi citra dengan morphology transformation hingga mendapatkan citra yang mirip dengan expected output pada gambar diatas!


```python
# import gambar
eye = io.imread('eye.png', as_gray=True)

# membuat strel
strel = square(30)

# melakukan opening
opened = opening(eye, strel)

# melakukan closing
closed = closing(opened, strel)

# show image
plt.figure(figsize=(15,10))
plt.subplot(1, 2, 1)
plt.title("Before")
plt.imshow(eye, cmap="gray")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title("After")
plt.imshow(closed, cmap="gray")
plt.axis("off")
plt.show()
```

### b. [10] Ceritakan mengapa Anda mengerjakan poin a dengan cara tersebut!

Karena, pertama lakukan opening untuk menghilangkan persegi panjang yang menghubungkan antara kedua persegi dan persegi dengan lingkaran. Kemudian lakukan closing untuk memenuhi kedua persegi.

## 2. [25] Aksara Jawa with a Hole

### a. [15] Lakukan berbagai percobaan manipulasi citra dengan morphology transformation hingga mendapatkan citra yang mirip dengan expected output pada gambar diatas!

```python
# import gambar
aksara_jawa = io.imread('aksara_jawa.png', as_gray=True)

# membuat strel
strel = disk(5)

# melakukan erosi
eroded = erosion(aksara_jawa, strel)

# melakukan pengurangan antara eroded dan gambar asli
res = eroded - aksara_jawa

# show image
plt.figure(figsize=(15,7))
plt.subplot(1, 2, 1)
plt.title("Before")
plt.imshow(aksara_jawa, cmap="gray")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title("After")
plt.imshow(res, cmap="gray")
plt.axis("off")
plt.show()
```

### b. [10] Ceritakan mengapa Anda mengerjakan poin a dengan cara tersebut!

Hasil dari eroded bentuknya lebih besar dibandingkan dengan gambar aslinya jadi ketika eroded dikurangi dengan gambar aslinya akan menghasilkan gambar tersebut.

## 3. [50] Image Preprocessing for Plate Number Recognition

### a. [40] Ceritakan bagaimana anda akan melakukan manipulasi terhadap citra mobil tersebut dengan menggunakan morphology transformation! Kemudian, lakukan manipulasi citra mobil tersebut dengan menggunakan morphology transformation sehingga didapatkan citra yang lebih baik.

Disini saya menggunakan bottom hat filter untuk mengekstrak detil-detil yang berwarna gelap yaitu plat nomor.

```python
# import gambar
small_car_1 = color.rgb2gray(io.imread('small_car_1.jpg'))
small_car_2 = color.rgb2gray(io.imread('small_car_2.jpg'))
small_car_3 = color.rgb2gray(io.imread('small_car_3.jpg'))
small_car_4 = color.rgb2gray(io.imread('small_car_4.jpg'))
small_car_5 = color.rgb2gray(io.imread('small_car_5.jpg'))

# membuat strel
strel = square(6)

# melakukan bottom hat filter untuk semua gambar
res_1 = black_tophat(small_car_1, strel)
res_2 = black_tophat(small_car_2, strel)
res_3 = black_tophat(small_car_3, strel)
res_4 = black_tophat(small_car_4, strel)
res_5 = black_tophat(small_car_5, strel)

# Show image 1
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image 1')
plt.imshow(small_car_1, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Bottom-Hat Filter 1')
plt.imshow(res_1, cmap='gray')
plt.axis('off')
plt.show()
# Show image 2
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image 2')
plt.imshow(small_car_2, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Bottom-Hat Filter 2')
plt.imshow(res_2, cmap='gray')
plt.axis('off')
plt.show()
# Show image 3
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image 3')
plt.imshow(small_car_3, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Bottom-Hat Filter 3')
plt.imshow(res_3, cmap='gray')
plt.axis('off')
plt.show()
# Show image 4
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image 4')
plt.imshow(small_car_4, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Bottom-Hat Filter 4')
plt.imshow(res_4, cmap='gray')
plt.axis('off')
plt.show()
# Show image 5
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Original Image 5')
plt.imshow(small_car_5, cmap='gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Bottom-Hat Filter 5')
plt.imshow(res_5, cmap='gray')
plt.axis('off')
plt.show()
```
